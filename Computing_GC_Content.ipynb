```python
def calc_GC(string):
    g=string.count('G')
    c=string.count('C')
    gc=g+c
    if len(string)==0:
        return 0
    percent_gc=100*gc/len(string)
    return percent_gc
    print(percent_gc)
    
def string_comp(file_name):
    seq='' #establishes the seq variable
    best=0 #establishes the baseline percentage to "beat"
    ID=''
    F=open(file_name, 'a') #opens the file in append mode
    F.write('\n>') #adds a > to the end of the file, to ensure that the final calc_GC is executed.
    F.close() #closes the file
    F=open(file_name) #opens the file in read mode
    for line in F: #reads the file line by line
        if line[0]=='>': #if the line starts with >, we first to determine the GC% of the previous sequence. 
            if calc_GC(seq)>best: #first, checks if there is a previously concatated string that the %GC can be calculated for, and determines if it is the highest percentage seen yet
                best=calc_GC(seq) #if it IS the highest % seen yet, saves it under the best variable
                bestID=ID #also saves the ID of the string with that highest %
            ID=line.strip() #keeps track of which string we are looking at currently. Gets reset with each new string.
            seq='' #resets the seq variable
        else:
            seq=seq+line.strip() #make the seq variable equal to the sequence following the >Title, plus whatever has come before it since the last >Title. This effectively combines the multiple lines for the sequences. 
    print('The string with the highest GC content is:',bestID)
    print('The GC content is:',best)
            
            
#calc_GC(' CCACCCTCGTGGTATGGCTAGGCATTCAGGAACCGGAGAACGCTTCAGACCAGCCCGGACTGGGAACCTGCGGGCAGTAGGTGGAAT')   
string_comp('rosalind_gc.txt')
```
