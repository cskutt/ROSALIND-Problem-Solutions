```python
def string_comp(file_name):
    seq='' 
    ID=''
    i=0
    a_list=[]
    c_list=[]
    g_list=[]
    t_list=[]
    A=''
    C=''
    G=''
    T=''
    final_string=''
    F=open(file_name) #opens the file in read mode
    for line in F: #reads the file line by line
        if line[0]=='>': #if the line starts with >, we now make sure that the previous line (the DNA sequence) gets computed
            
            #Loop that reads the sequence, one nucleotide at a time. 
            #Then, it adds the value of the index that each nucleotide is found at to a list/string
            #The result will be 4 different strings, one for each type of nucleotide (a, c, g, t)
            #The strings will have a list of numbers of each position that that type of nucleotide is found at
            
            #print(seq)
            #print(line)
            i=0
            while i < len(seq):
                if seq[i] == 'A':
                    a_list.append(i)
                if seq[i] == 'C':
                    c_list.append(i)
                if seq[i] == 'G':
                    g_list.append(i)
                if seq[i] == 'T':
                    t_list.append(i)
                i=i+1
            matrix_length=(len(seq)) #quickly grabs the length of the sequence for later use when creating the needed arrays
            seq=''
        else:
            seq=seq+line.strip() #make the seq variable equal to the sequence following the >Title, plus whatever has come before it since the last >Title. This effectively combines the multiple lines for the sequences.     


    #This step creates python lists for each nucleotide. 
    #Uses the matrix_length variable created earlier to ensure that the arrays are as many values long as the DNA sequences 
    #Then, I can edit the lists at specific indices to add up the nucleotide list values from earlier
    As=[]
    Cs=[]
    Gs=[]
    Ts=[]
    i=0
    while i < matrix_length:
        As.append(0)
        Cs.append(0)
        Gs.append(0)
        Ts.append(0)
        i=i+1

    
    #This step assigns the nucleotide positions to the python lists.
    #It goes through the nucleotide position string one at a time, and then adds 1 to the python list position that corresponds to the nucleotide position value. 
    #For example, an A nucleotide found at index 5 will have a value of '5' in the nucleotide list. 
    #This will +1 to the As list at index 5
    #The resulting list will list how many of each nucleotide were found at each index. 
    i=0
    while i < len(a_list):
        if a_list[i]==' ':
            i=i+1
            continue
        value=int(a_list[i])
        As[value]=As[value]+1
        i=i+1
    i=0
    while i < len(c_list):
        if c_list[i]==' ':
            i=i+1
            continue
        value=int(c_list[i])
        Cs[value]=Cs[value]+1
        i=i+1
    i=0
    while i < len(g_list):
        if g_list[i]==' ':
            i=i+1
            continue
        value=int(g_list[i])
        Gs[value]=Gs[value]+1
        i=i+1
    i=0
    while i < len(t_list):
        if t_list[i]==' ':
            i=i+1
            continue
        value=int(t_list[i])
        Ts[value]=Ts[value]+1
        i=i+1
    
    
    #Finally, this step compares the lists to determine which nucleotide is most present at any given position and create the consensus string
    #If the A nucleotide is the most present at a given position, it will add an "A" to the consensus string
    i=0
    while i < len(As):
        if As[i] >= Cs[i] and As[i]>=Gs[i] and As[i]>=Ts[i]:
            final_string=final_string+'A'
        elif Cs[i] >= As[i] and Cs[i]>=Gs[i] and Cs[i]>=Ts[i]:
            final_string=final_string+'C'
        elif Gs[i] >= Cs[i] and Gs[i]>=As[i] and Gs[i]>=Ts[i]:
            final_string=final_string+'G'
        elif Ts[i] >= Cs[i] and Ts[i]>=Gs[i] and Ts[i]>= As[i]:
            final_string=final_string+'T'
        else:
            print('wah')
        i=i+1
        
    i=0
    while i<len(As):
        A=A+str(As[i])+' '
        C=C+str(Cs[i])+' '
        G=G+str(Gs[i])+' '
        T=T+str(Ts[i])+' '
        i=i+1
    print(final_string)
    print('A:',A)
    print('C:',C)
    print('G:',G)
    print('T:',T)
            
         
            
string_comp('rosalind_cons.txt')
```
