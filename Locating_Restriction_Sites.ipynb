```python
def is_rev_palindrome(string):
    #Create a dictionary that contains the pair information for nucleotides
    pairs={"A":"T","G":"C","T":"A","C":"G"}
    
    #Create a string to keep track of the reverse pairs
    reverse_pairs=''
    
    #For each letter in the original string (reading back to front),
    #add the corresponding pair to the reverse_pairs string
    for i in range(len(string) -1,-1,-1):
        reverse_pairs+=pairs[string[i]]
        
    #Check whether the reverse string is equal to the original string    
    return string == reverse_pairs


        
def locate(string):
    #Create a list to hold the reverse palindromes and their positions
    rev_pals=[]
    #Loop through the dna string from the start to the end
    for i in range(len(string)):
        #Since we are looking for rev pals between 4 and 12 characters,
        #We set n to be between 4 and 13
        for n in range(4,13):
            if i+n>len(string):
                continue
            if is_rev_palindrome(string[i:i+n]):
                rev_pals.append((i+1,n))
    
    for position,length in rev_pals:
        print(position,length)
            


locate('GCGCTCATTAAGATCGCGGCATTGAATCGTGGAAACTGCGGCAAGCAGGATCACAAGCATCTGACGTTGACCCCGGTCGTTACTGTGAGATGTGGTCCGTGTTCAATTGCCATCCCGCGTTTTACGCAGAAGCGTACTTACTACGAGGCGCGACGCCATCCATCTTACTAGGTGTAACGAATGTATTTGCTGACATTCAGGTTAACATACGGACAACAATCTATGCAGCGGCACGTGCGCTCTGACGAACCCGAGAGCCGGGGCTTACCTCTCGCCTACGAATCCGCCGAGACATGAACTGTCGGATAACTACCCCAATTCTTCACTGGGTACAATCAGATGGTTGTCCAGGACCATATCTTGAACCACGTGACTCGCCAACCTCCCGTTCAGAAGTTCAAGGTCTTCACAGAGAATTGTCAACTACTAGAAGGTGTAGGCTCCAGGCTCATGTATATAAGGGAAACCACTGATATCATTCGGCGAGGCCTCGCCTGCTTGTTCGGATGTTGCCTATCTCGGAACAGGTGTTCCTACAAATTGATTCACCCCGGAAGCCGGGTTAGAATATAAACGTTCCTTTTAACACATGTTTGGCATGCGTGCGGTGGAGTCGATGTCCGATATATAACGCAACGCGGCAAGAAACTTTTAGCCGCTGTAGCCACTCCCGTTAAGTTCCTCACGTTACTACGAGCCTGTCTTGGGTAAATCTCCTGTCACCGCTCACGTAGAAAGGCAATCAGATTTGGGAATCTTTAAGAAATCGGTTAAGTATCGTTACTTATCCCCAAATTAGATGATGTGACGTCCATGACCAGACGGGCTAAGATAAACCCATGGTGGTTGCCTCAGCCATCAAAGCGAACTTGCTATCCGCAGCCTATATACTGGAATCTCTCTCTGAACCGGGCTGTGTTACTTACGGCGCGTTACTCCGTCGCCATGATCTTGGATCATAATACAATTTTCTATA')
```
